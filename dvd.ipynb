```python
import pandas as pd
import numpy as np
df = pd.read_csv('voting-survey_response (2).csv')
```

```python
df.drop('Q1', axis=1, inplace=True)
df.replace(-1, np.nan, inplace=True)
```

```python
not_voted = df[(df['Q29_1'] == 1) |  (df['Q29_2'] == 1) |(df['Q29_3'] == 1) |(df['Q29_4'] == 1) |(df['Q29_5'] == 1) |(df['Q29_6'] == 1) | (df['Q29_7'] == 1) |(df['Q29_8'] == 1) | (df['Q29_9'] == 1) |(df['Q29_10'] == 1)]

voted = df[(df['Q29_1'] != 1) &   (df['Q29_2'] != 1) & (df['Q29_3'] != 1) & (df['Q29_4'] != 1) & (df['Q29_5'] != 1) & (df['Q29_6'] != 1) &  (df['Q29_7'] != 1) & (df['Q29_8'] != 1) &  (df['Q29_9'] != 1) & (df['Q29_10'] != 1)]
```

```python
q2_v=not_voted[['RespId'	,'weight',	'age',	'Education',	'race',	'gender',	'income_cat',	'voter_category' ,'Q2_1','Q2_2'	,'Q2_3'	,'Q2_4', 'Q2_5','Q2_6'	,'Q2_7'	,'Q2_8', 'Q2_9','Q2_10'	]]

lables=[
'Union Voting',
'Community Meetings',
'Stay Informed',
'Community Emblem',
'Community Events',
'Know Pledge & Rules',
'Support Worker Rights',
'Respect Dissent',
'Belief in God',
'Protest Injustice'
]


for index, j in enumerate(lables):
    print(index+1,j)
    q2_v.rename(columns={f'Q2_{index+1}': j}, inplace=True)

q2_v.replace(1, 'Very important', inplace=True)
q2_v.replace(2, 'Somewhat important', inplace=True)
q2_v.replace(3, 'Not so important', inplace=True)
q2_v.replace(4, 'Not at all important', inplace=True)

# voting
q2df=q2_v[lables]
q2df.apply(pd.Series.value_counts)


HispanicQ2=q2_v[q2_v['race']== 'Hispanic'][lables]
HispanicQ2.apply(pd.Series.value_counts).to_csv(f'{voted_folder}/q2_pivot_hispanic.csv')

eduQ2=q2_v[q2_v['Education']== 'High school or less'][lables]
eduQ2.apply(pd.Series.value_counts).to_csv(f'{voted_folder}/q2_edu.csv')

ageQ2=q2_v[q2_v['age']<40][lables]
ageQ2.apply(pd.Series.value_counts).to_csv(f'{voted_folder}/q2_age.csv')

incomeQ2=q2_v[q2_v['income_cat'] == 'Less than $40k'][lables]
incomeQ2.apply(pd.Series.value_counts).to_csv(f'{voted_folder}/q2_income.csv')
```
